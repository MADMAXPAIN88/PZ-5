```python
def calculator(x, y, p):
    if p == "+":
      return x + y
    elif p == "-":
      return x - y
    elif p == "*":
      return x * y
    elif p == "/":
      return x / y

p = input("Введите операцию (+, -, *, /): ")
x = int(input("Введите первое число: "))
y = int(input("Введите второе число: "))

print(f'Итог: {calculator(x, y, p)}')
```
